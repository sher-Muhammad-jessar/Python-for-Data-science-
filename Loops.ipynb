```python
# range function in for loop 
# range(start size, stop,stepsize)
# method 1 range(5) only stop size it start from 0 and increase by 1 (by default)
# method 2  range(5,10) 5-> start,  10->end increment by 1 (by default)
# method 3 range(2,10,3) 2-> start, 10-> stop , 3-> increment by 


for el in range(10):
    print(el)

print("method 2")

for el in range(2,10):
     print(el)
print("method 3")
for el in range(1,10,2):
    print(el)
```
